```python
from matplotlib import pyplot as plt
Arabic = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
English = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'] 
plt.scatter(Arabic, English)
# label each point
for x1, x2, x3 in zip(labels, Arabic, English): 
  plt.annotate(x1,xy=(x2, x3), xytext=(5, -5), textcoords='offset points')
plt.title("lab 3 plooot") 
plt.xlabel("# Arabic")
plt.ylabel("matploooot")
plt.show()
```
